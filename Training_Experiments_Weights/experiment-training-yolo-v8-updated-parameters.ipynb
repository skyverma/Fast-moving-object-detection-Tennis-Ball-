```python
import os

# Disable Weights & Biases
os.environ['WANDB_MODE'] = 'disabled'
```

```python
# Import the YOLO class from ultralytics
from ultralytics import YOLO

# # Load the pretrained YOLOv8 model
# model = YOLO('yolov8x.pt')
# from ultralytics import YOLO

# Load the pretrained YOLOv8 model
model = YOLO('yolov8x.pt')

# Train the model using the custom dataset
model.train(data='/kaggle/input/dataset-for-initial-training/data.yaml',
            epochs=100,           # Number of epochs
            imgsz=640,            # Image size
            batch=8,              # Reduce batch size to avoid GPU memory issues
            patience=5,           # Early stopping if no improvement
            lr0=1e-3,             # Learning rate, lower if unstable
            optimizer='Adam',     # Adam optimizer for better performance on small datasets
            augment=True,         # Enable augmentations
            hsv_h=0.015,          # Color augmentations (hue, saturation, brightness)
            hsv_s=0.7,
            hsv_v=0.4,
            fliplr=0.5,           # Horizontal flip
            mosaic=1.0,           # Use mosaic augmentation
            cache=True)           # Cache images for faster training
```

```python
# Evaluate the model on the test set
results = model.val()
```

```python
# Perform inference on a new image
results = model.predict(source='/kaggle/input/dataset-for-initial-training/test/images', save=True)
```
